```python
import matplotlib.pyplot as plt
import numpy as np
import os
#############################################        M=2,K=2                      ######################################################
fig = plt.figure()  #定义新的三维坐标轴
ax3 = plt.axes(projection='3d')
#定义三维数据
small_number = 0.01 
x1 = np.arange(-1,-small_number,0.01)
x2 = np.arange(small_number,1,0.01)
xx = np.concatenate([x1, x2])

y1 = np.arange(-1,-small_number,0.01)
y2 = np.arange(small_number,1,0.01)
yy = np.concatenate([y1, y2])

X, Y = np.meshgrid(xx, yy)

a1 = X 
a2= 0 - a1
b1 = Y 
b2= 0 - b1

Z = (a1+b1)*(a1+b1)+(a2+b2)*(a2+b2)
#作图
ax3.set_xlabel('Diversity')
ax3.set_ylabel('Teachers Error')
ax3.set_zlabel('Student Error')
ax3.plot_surface(0.5*(2*a1*b1+2*a2*b2),0.5*(a1*a1+a2*a2+b1*b1+b2*b2),0.5*Z,rstride = 1, cstride = 1,cmap='rainbow')
ax3.view_init(-120, -60)
if not os.path.isdir('picture/'):
    os.mkdir('picture/')
plt.savefig('picture/M2K2.jpg')
plt.show()
```

```python
#############################################        M=2,K=3                      ######################################################
fig = plt.figure()  #定义新的三维坐标轴
ax3 = plt.axes(projection='3d')
#定义三维数据
small_number = 0.01 
x1 = np.arange(-1,-small_number,0.05)
x2 = np.arange(small_number,1,0.05)
xx = np.concatenate([x1, x2])
fineness = 1600
X, Y, W, V = np.meshgrid(xx, xx, xx, xx)

X1 = X.reshape(fineness,fineness)
X2 = Y.reshape(fineness,fineness)
X3 = 0 - X1 - X2
Y1 = W.reshape(fineness,fineness)
Y2 = V.reshape(fineness,fineness)
Y3 = 0 - Y1 - Y2

j = 0
for i in range(len(X3)):
    condition = np.where((X3[i]>-1)&(X3[i]<1) & (Y3[i]>-1)&(Y3[i]<1))
    
    A1 = X1[i][condition]
    A2 = X2[i][condition]
    A3 = X3[i][condition]
    B1 = Y1[i][condition]
    B2 = Y2[i][condition]
    B3 = Y3[i][condition]
    
    if len(A1)==0:
        continue
    else:
        j = j + 1
        
    if j == 1:
        a1 = np.expand_dims(A1, axis=0)
        a2 = np.expand_dims(A2, axis=0)
        a3 = np.expand_dims(A3, axis=0)
        b1 = np.expand_dims(B1, axis=0)
        b2 = np.expand_dims(B2, axis=0)
        b3 = np.expand_dims(B3, axis=0)
    else:
        a1 = np.concatenate([a1, np.expand_dims(A1, axis=0)], 0)
        a2 = np.concatenate([a2, np.expand_dims(A2, axis=0)], 0)
        a3 = np.concatenate([a3, np.expand_dims(A3, axis=0)], 0)
        b1 = np.concatenate([b1, np.expand_dims(B1, axis=0)], 0)
        b2 = np.concatenate([b2, np.expand_dims(B2, axis=0)], 0)
        b3 = np.concatenate([b3, np.expand_dims(B3, axis=0)], 0)

Z = (a1+b1)*(a1+b1)+(a2+b2)*(a2+b2)+(a3+b3)*(a3+b3)
#作图
ax3.set_xlabel('Diversity')
ax3.set_ylabel('Teachers Error')
ax3.set_zlabel('Student Error')
ax3.plot_surface(0.33*(2*a1*b1+2*a2*b2+2*a3*b3),0.33*(a1*a1+a2*a2+a3*a3+b1*b1+b2*b2+b3*b3),0.33*Z,rstride = 1, cstride = 1,cmap='rainbow')
ax3.view_init(-120, -60)
if not os.path.isdir('picture/'):
    os.mkdir('picture/')
plt.savefig('picture/M2K3.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

#############################################        M=6,K=3                      ######################################################
fig = plt.figure()  #定义新的三维坐标轴
ax3 = plt.axes(projection='3d')
#定义三维数据
small_number = 0.01
x1 = np.arange(-1,-small_number,0.5)
x2 = np.arange(small_number,1,0.5)
xx = np.concatenate([x1, x2])
fineness = 4096
X, Y, W, V, U, S, O, P, Q, R, T, L = np.meshgrid(xx, xx, xx, xx, xx, xx, xx, xx, xx, xx, xx, xx)

X1 = X.reshape(fineness,fineness)
X2 = Y.reshape(fineness,fineness)
X3 = 0 - X1 - X2
Y1 = W.reshape(fineness,fineness)
Y2 = V.reshape(fineness,fineness)
Y3 = 0 - Y1 - Y2
W1 = U.reshape(fineness,fineness)
W2 = S.reshape(fineness,fineness)
W3 = 0 - W1 - W2
M1 = O.reshape(fineness,fineness)
M2 = P.reshape(fineness,fineness)
M3 = 0 - M1 - M2
N1 = Q.reshape(fineness,fineness)
N2 = R.reshape(fineness,fineness)
N3 = 0 - N1 - N2
T1 = T.reshape(fineness,fineness)
T2 = L.reshape(fineness,fineness)
T3 = 0 - T1 - T2

j = 0
for i in range(len(X3)):
    condition = np.where((X3[i]>-1)&(X3[i]<1) & (Y3[i]>-1)&(Y3[i]<1) & (W3[i]>-1)&(W3[i]<1)& (M3[i]>-1)&(M3[i]<1) \
                         & (N3[i]>-1)&(N3[i]<1)& (T3[i]>-1)&(T3[i]<1))
    
    A1 = X1[i][condition]
    A2 = X2[i][condition]
    A3 = X3[i][condition]
    B1 = Y1[i][condition]
    B2 = Y2[i][condition]
    B3 = Y3[i][condition]
    C1 = W1[i][condition]
    C2 = W2[i][condition]
    C3 = W3[i][condition]
    D1 = M1[i][condition]
    D2 = M2[i][condition]
    D3 = M3[i][condition]
    E1 = N1[i][condition]
    E2 = N2[i][condition]
    E3 = N3[i][condition]
    F1 = T1[i][condition]
    F2 = T2[i][condition]
    F3 = T3[i][condition]
    
    if len(A1)==0:
        continue
    else:
        A1= np.pad(A1,(0,fineness-len(A1)),'constant', constant_values=(A1[0],A1[0]))
        A2= np.pad(A2,(0,fineness-len(A2)),'constant', constant_values=(A2[0],A2[0]))
        A3= np.pad(A3,(0,fineness-len(A3)),'constant', constant_values=(A3[0],A3[0]))
        B1= np.pad(B1,(0,fineness-len(B1)),'constant', constant_values=(B1[0],B1[0]))
        B2= np.pad(B2,(0,fineness-len(B2)),'constant', constant_values=(B2[0],B2[0]))
        B3= np.pad(B3,(0,fineness-len(B3)),'constant', constant_values=(B3[0],B3[0]))
        C1= np.pad(C1,(0,fineness-len(C1)),'constant', constant_values=(C1[0],C1[0]))
        C2= np.pad(C2,(0,fineness-len(C2)),'constant', constant_values=(C2[0],C2[0]))
        C3= np.pad(C3,(0,fineness-len(C3)),'constant', constant_values=(C3[0],C3[0]))
        D1= np.pad(D1,(0,fineness-len(D1)),'constant', constant_values=(D1[0],D1[0]))
        D2= np.pad(D2,(0,fineness-len(D2)),'constant', constant_values=(D2[0],D2[0]))
        D3= np.pad(D3,(0,fineness-len(D3)),'constant', constant_values=(D3[0],D3[0]))
        E1= np.pad(E1,(0,fineness-len(E1)),'constant', constant_values=(E1[0],E1[0]))
        E2= np.pad(E2,(0,fineness-len(E2)),'constant', constant_values=(E2[0],E2[0]))
        E3= np.pad(E3,(0,fineness-len(E3)),'constant', constant_values=(E3[0],E3[0]))
        F1= np.pad(F1,(0,fineness-len(F1)),'constant', constant_values=(F1[0],F1[0]))
        F2= np.pad(F2,(0,fineness-len(F2)),'constant', constant_values=(F2[0],F2[0]))
        F3= np.pad(F3,(0,fineness-len(F3)),'constant', constant_values=(F3[0],F3[0]))
        j = j + 1
        
    if j == 1:
        a1 = np.expand_dims(A1, axis=0)
        a2 = np.expand_dims(A2, axis=0)
        a3 = np.expand_dims(A3, axis=0)
        b1 = np.expand_dims(B1, axis=0)
        b2 = np.expand_dims(B2, axis=0)
        b3 = np.expand_dims(B3, axis=0)
        c1 = np.expand_dims(C1, axis=0)
        c2 = np.expand_dims(C2, axis=0)
        c3 = np.expand_dims(C3, axis=0)
        d1 = np.expand_dims(D1, axis=0)
        d2 = np.expand_dims(D2, axis=0)
        d3 = np.expand_dims(D3, axis=0)
        e1 = np.expand_dims(E1, axis=0)
        e2 = np.expand_dims(E2, axis=0)
        e3 = np.expand_dims(E3, axis=0)
        f1 = np.expand_dims(F1, axis=0)
        f2 = np.expand_dims(F2, axis=0)
        f3 = np.expand_dims(F3, axis=0)
    else:
        a1 = np.concatenate([a1, np.expand_dims(A1, axis=0)], 0)
        a2 = np.concatenate([a2, np.expand_dims(A2, axis=0)], 0)
        a3 = np.concatenate([a3, np.expand_dims(A3, axis=0)], 0)
        b1 = np.concatenate([b1, np.expand_dims(B1, axis=0)], 0)
        b2 = np.concatenate([b2, np.expand_dims(B2, axis=0)], 0)
        b3 = np.concatenate([b3, np.expand_dims(B3, axis=0)], 0)
        c1 = np.concatenate([c1, np.expand_dims(C1, axis=0)], 0)
        c2 = np.concatenate([c2, np.expand_dims(C2, axis=0)], 0)
        c3 = np.concatenate([c3, np.expand_dims(C3, axis=0)], 0)
        d1 = np.concatenate([d1, np.expand_dims(D1, axis=0)], 0)
        d2 = np.concatenate([d2, np.expand_dims(D2, axis=0)], 0)
        d3 = np.concatenate([d3, np.expand_dims(D3, axis=0)], 0)
        e1 = np.concatenate([e1, np.expand_dims(E1, axis=0)], 0)
        e2 = np.concatenate([e2, np.expand_dims(E2, axis=0)], 0)
        e3 = np.concatenate([e3, np.expand_dims(E3, axis=0)], 0)
        f1 = np.concatenate([f1, np.expand_dims(F1, axis=0)], 0)
        f2 = np.concatenate([f2, np.expand_dims(F2, axis=0)], 0)
        f3 = np.concatenate([f3, np.expand_dims(F3, axis=0)], 0)

Z = (a1+b1+c1+d1+e1+f1)*(a1+b1+c1+d1+e1+f1)+(a2+b2+c2+d2+e2+f2)*(a2+b2+c2+d2+e2+f2)+(a3+b3+c3+d3+e3+f3)*(a3+b3+c3+d3+e3+f3)
#作图
ax3.set_xlabel('Diversity')
ax3.set_ylabel('Teachers Error')
ax3.set_zlabel('Student Error')
ax3.plot_surface(0.33*(2*a1*b1+2*a1*c1+2*a1*d1+2*b1*c1+2*b1*d1+2*c1*d1+2*a1*e1+2*b1*e1+2*c1*e1+2*d1*e1+2*a1*f1+2*b1*f1+2*c1*f1+2*d1*f1+2*e1*f1+ \
                 2*a2*b2+2*a2*c2+2*a2*d2+2*b2*c2+2*b2*d2+2*c2*d2+2*a2*e2+2*b2*e2+2*c2*e2+2*d2*e2+2*a2*f2+2*b2*f2+2*c2*f2+2*d2*f2+2*e2*f2+ \
                 2*a3*b3+2*a3*c3+2*a3*d3+2*b3*c3+2*b3*d3+2*c3*d3+2*a3*e3+2*b3*e3+2*c3*e3+2*d3*e3+2*a3*f3+2*b3*f3+2*c3*f3+2*d3*f3+2*e3*f3), \
                 0.33*(a1*a1+a2*a2+a3*a3+b1*b1+b2*b2+b3*b3+c1*c1+c2*c2+c3*c3+d1*d1+d2*d2+d3*d3+e1*e1+e2*e2+e3*e3+f1*f1+f2*f2+f3*f3),0.33*Z,rstride = 1, cstride = 1,cmap='rainbow')
ax3.view_init(-120, -60)
if not os.path.isdir('picture/'):
    os.mkdir('picture/')
plt.savefig('picture/M6K3.jpg')
plt.show()
```
